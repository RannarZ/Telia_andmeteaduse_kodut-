# Telco Customer Churn #



## Importimised ##

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

## Andmete eeltöötlus ##

```python
#Loeme andmed sisse
telco = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
#Teame kohe, et see pole relevantne tulp, kuna see on lihtsalt identifikaator
telco = telco.drop(['customerID'], axis=1)
```

```python
telco.head()
```

```python
telco['Contract'].value_counts()
```

```python
#Joonistelt on näha, et TotalCharges tulp on vale väärtusega ning SeniorCitizen väärtused on 1 ja 0 booleanide asemel
#Muudame õigeteks andmetüüpideks ehk TotalCharges arvuks ning SeniorCitizen booleaniks, et saaks graafe teha õigesti
telco['TotalCharges'] = pd.to_numeric(telco['TotalCharges'], errors='coerce')
telco['SeniorCitizen'] = telco['SeniorCitizen'].astype(bool)
#Eemaldame nan väärtused
telco = telco.dropna()
```

```python
def joonised_nominaal(andmed):
    for column in andmed.columns:
        if column != 'Churn' and (andmed[column].dtypes == 'object' or andmed[column].dtypes == 'bool'):
            #Andmete grupeerimine
            grouped = andmed.groupby(column)['Churn'].value_counts().unstack(fill_value=0)
            #Joonise loomine
            grouped.plot(kind='bar', stacked=False)  # Use stacked=True for stacked bars
            plt.xlabel(column)
            plt.ylabel('Esinemiste arv')
            plt.title(f'Ettevõttest lahkunute ja mitte lahkunute arvu suhe \'{column}\' tulbaga')
            plt.legend(title=column)
            plt.show()
joonised_nominaal(telco)
```

```python
def joonised_arvud(andmed):
    for column in andmed.columns:
        if andmed[column].dtypes == 'int64' or andmed[column].dtypes == 'float64':
            #Andmete grupeerimine
            true_väärtus = andmed.loc[andmed['Churn'] == 'Yes', column]
            false_väärtus = andmed.loc[andmed['Churn'] == 'No', column]
            #Joonise loomine
            plt.hist(true_väärtus, alpha=0.5, label='Yes', bins=10) 
            plt.hist(false_väärtus, alpha=0.5, label='No', bins=10) 
            plt.xlabel(column)
            plt.ylabel('Churn')
            plt.title(f'{column}')
            plt.legend()
            plt.show()
joonised_arvud(telco)
```

```python
#Andmete eeltöötlus
def eeltöötlus(andmed):
    
    #Puuduvate väärtustega ei ole vaja tegeleda, kuna neid ei ole
    #print(data.isna().sum())
    töödeldud_andmed = andmed.copy()
    töödeldud_andmed = töödeldud_andmed.drop(['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'PaperlessBilling'], axis=1)
    ennustatav = töödeldud_andmed['Churn']
    töödeldud_andmed = töödeldud_andmed.drop(['Churn'], axis=1)

    #Leiame kõik kahe väärtusega tulbad
    kahesed_tulbad = []
    for column in töödeldud_andmed.columns:
        if len(töödeldud_andmed[column].value_counts()) == 2:
            kahesed_tulbad.append(column)
    print(kahesed_tulbad)

    #Siia saab lisada onehot vektorite jaotused ning Churn tulba kui seda on
    andmestikud = []

    #Kõikidel ülejäänutel kaheväärtuselistel tulpadel muudame vastavalt other_values'i järgi
    other_values = {'Yes' : True, 'No' : False}
    for column in kahesed_tulbad[1:]:
        töödeldud_andmed[column] = töödeldud_andmed[column].map(other_values)

    #Kuna contract tulbas on ordinaalsed tunnused (aeg), siis teeme sellele tulbale numbrilise kodeeringu
    contract_values = {'Month-to-month' : 1, 'One year' : 2, 'Two year' : 3}
    töödeldud_andmed['Contract'] = töödeldud_andmed['Contract'].map(contract_values)

    vektor_nimed = []
    #Teeme kõikidest ülejäänutest nominaalsetest tulpadest onehot vektorid
    for column in töödeldud_andmed.columns:
        if töödeldud_andmed[column].dtype == 'object':
            vektor_nimed.append(column)
            onehot_tulp = pd.get_dummies(töödeldud_andmed[column])
            #Kontrollime, kas tulba nimi on Yes või No ning kui on, siis lisame algusesse algse tulba nime
            #Seda selleks, et hiljem kaotada ära tulpade ühildamisviga, kus on mitu sama nime tulpadel.
            for tulp in onehot_tulp.columns:
                if tulp == 'Yes' or tulp == 'No':
                    onehot_tulp.columns = onehot_tulp.columns.str.replace(tulp, f'{column} {tulp}')
            andmestikud.append(onehot_tulp)
            
    #Lisame onehot vektorite tulbad andmetabelisse
    for onehot in andmestikud:
        töödeldud_andmed = töödeldud_andmed.join(onehot)

    #Eemaldame algsed tulbad, millest sai tehtud onehotid
    töödeldud_andmed = töödeldud_andmed.drop(vektor_nimed, axis=1)
    
    
    #Kui tekkis tühje väärtuseid, siis eemaldame tabelist
    #print(töödeldud_andmed.isna().sum())
    töödeldud_andmed = töödeldud_andmed.dropna()    
    return töödeldud_andmed, ennustatav
```

```python
andmed, ennustatav = eeltöötlus(telco)

#Eraldame treening-ja testandmed
X_train, X_test, y_train, y_test = train_test_split(andmed, ennustatav, test_size=0.2)

#andmed.head()
```

## Mudeli treenimine ##

```python
mudel = GradientBoostingClassifier()
mudel.fit(X_train, y_train)
```

## Mudeli testimine ##

```python
ennustused = mudel.predict(X_test)
täpsus = accuracy_score(ennustused, y_test)
print(f"Mudeli täpsus on: {täpsus.round(2)}.")
```

## Väike analüüs ##

Eeltöötlusel otsustasin graafikute põhjal, mis näitasid suhet ettevõttest lahkunute ning mittelahkunute vahel valitud tulbal, mis tulpi eemaldada. Eemaldasin tulbad, millel oli näha, et erinevatel väärtustel ei ole lahkunute ning mitte lahkunute suhe väga erinev. On näha, et lahkumise tõenäosus oli väga suur näiteks juhul, kui arveid maksti elektrooniliselt ning lepingud olid kuupõhised. Sammuti oli lahkumine võrdlemisi kõrge, kui puudus seadme kaitse, veebipõhine varundamine, tehniline tugi ning veebipõhine turvalisus.

Järgmiseks leidsin kõik kahe väärtusega tulbad ning muutsin need boolean tüüpi, et saaks andmete tabeli mudelile ette anda. Järgmiseks muutsin kõik nominaalsete tunnuste ning rohkem kui kahe väärtusega tulbad onehot vektoriteks. Tulp 'Contract' oli aga ordinaalne ehk tollele tegin numbrilise kodeerimise.

Kuna kasutame ühte kindlat andmete tabelit, siis kodeerisin kõik andmed korraga onehot vektoriteks. Kui testimise andmed tuleksid mõnest muust andmetabelist, siis peaks treening-ja testandmete eeltöötluse tegema eraldi ning hiljem vaja minevad tulbad ning väärtused ühendada (et saada sama struktuur lõpuks).

Klassifitseerijaks valisin GradientBoostingClassifieri, kuna see andis kõige paremaid tulemusi.
